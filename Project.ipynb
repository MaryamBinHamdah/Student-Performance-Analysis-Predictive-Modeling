##**6- Create a script to develop a Python function for descriptive statistics. The input for the function should be the sample and the field to perform the descriptive statistics.**   

from matplotlib.colors import Normalize

````python

def Desc_stat(sample, var):
    mean = sample[var].mean()
    median = sample[var].median()
    mode = sample[var].mode().iloc[0]
    minimum = sample[var].min()
    maximum = sample[var].max()
    range_value = maximum - minimum
    std_dev = sample[var].std()
    variance = sample[var].var()
    skew = sample[var].skew()
    kurtosis = sample[var].kurt()
    count = sample[var].count()

    Quart = [
        sample[var].quantile(0),
        sample[var].quantile(0.25),
        sample[var].quantile(0.50),
        sample[var].quantile(0.75),
        sample[var].quantile(1),
        sample[var].quantile(0.75) - sample[var].quantile(0.25)
    ]

    summary = {
        "Average": mean,
        "Median": median,
        "Mode": mode,
        "Minimum": minimum,
        "Maximum": maximum,
        "Range": range_value,
        "Std Dev": std_dev,
        "Variance": variance,
        "Skewness": skew,
        "Kurtosis": kurtosis,
        "Minimum Quartile": Quart[0],
        "25th Percentile": Quart[1],
        "50th Percentile": Quart[2],
        "75th Percentile": Quart[3],
        "Maximum Quartile": Quart[4],
        "Inter Quartile Range": Quart[5],
        "Count": count
    }

    return summary
````

##**11 - hypothesis test to find the correlation (Chi-Square Test)**

Perform Correlation test -Chi-square Test

A. NULL HYPOTHESIS (H0): gender is independent of test preparation course

B. ALTERNATIVE HYPOTHESIS (H1): gender is dependent on test preparation course


````python
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_data = pd.crosstab(students_performance['gender'], students_performance['test preparation course'])

# Perform Chi-Square Test
r, p, dof, expected = chi2_contingency(contingency_data)

# Print results
print('r=%.3f, p=%.3f' % (r, p))

if p > 0.05:
    print('independent categories')  # No significant association
else:
    print('dependent categories')  # Significant association
````

##**12- Assess the performance of the dependent variable to know whether the sample is representative of the normal population by a one-sample t-test.**

Purpose: To find the average (mean) writing score for female students is the same as the average of all students writing scores.
Assumption: Normally distributed

A. NULL HYPOTHESIS (H0): The average writing score for female students is the same as the average writing score for all students.

B. ALTERNATIVE HYPOTHESIS (H1): The average writing score for female students is not the same as the average writing score for all students.

````python
# import libraries
from scipy.stats import shapiro, ttest_1samp

# Extract writing scores for all students
writing_scores = students_performance['writing score']
print("Mean of writing scores:", writing_scores.mean())

# Extract writing scores for female students
writing_female = students_performance[students_performance['gender'] == 'female']['writing score']

# Check for normality of writing scores for female students using Shapiro-Wilk Test
r, p = shapiro(writing_female)
print('Shapiro-Wilk Test: r=%.2f, p=%.30f' % (r, p))

if p > 0.05:
    print('Normal distribution')
else:
    print('Not a normal distribution')

# Perform One-Sample T-Test on female students' writing scores to compare with overall mean
test_stat, p_value = ttest_1samp(writing_female, writing_scores.mean())

print(f"T-test Result: Statistic={test_stat:.2f}, p-value={p_value:.3f}")

# conclusion
if p_value > 0.05:
    print("As the p-value is more than 0.05, we accept the null hypothesis and reject the alternative hypothesis.")
else:
    print("As the p-value is less than 0.05, we reject the null hypothesis and accept the alternative hypothesis.")
````

## **15- Predict the value of the dependent variable from the different classifier such as Logistic Regression, KNN, Naïve-Bayes and Decision Tree.**

##**16- Evaluate the performance of each model using confusion matrix and accuracy and identify the best fit classifier for the chosen dataset.**

````python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

# Load the dataset
data = pd.read_csv('StudentsPerformance.csv')

# Create total score and then remove it (as per professor's notes)
data['total'] = data['math score'] + data['reading score'] + data['writing score']

# Create target variable (pass/fail) based on threshold of 60% of total possible (300)
threshold = 180  # 60% of 300
data['pass_fail'] = np.where(data['total'] >= threshold, 1, 0)  # 1 = pass, 0 = fail

# Now drop the total column as instructed
data = data.drop(columns=['total'])

# Select only reading and writing scores as independent variables (as per requirements)
X = data[['reading score', 'writing score']].values
y = data['pass_fail'].values

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

# Feature Scaling
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
````

** italicized textImplement Classification Models**

Logistic Regression


````python
# Logistic Regression
logistic_classifier = LogisticRegression(random_state=0)
logistic_classifier.fit(X_train, y_train)
y_pred_logistic = logistic_classifier.predict(X_test)

# Evaluation
cm_logistic = confusion_matrix(y_test, y_pred_logistic)
accuracy_logistic = accuracy_score(y_test, y_pred_logistic)
print("Logistic Regression:")
print("Confusion Matrix:\n", cm_logistic)
print("Accuracy:", accuracy_logistic)
print(classification_report(y_test, y_pred_logistic))
````

K-Nearest Neighbors (KNN)


````python
# KNN
knn_classifier = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
knn_classifier.fit(X_train, y_train)
y_pred_knn = knn_classifier.predict(X_test)

# Evaluation
cm_knn = confusion_matrix(y_test, y_pred_knn)
accuracy_knn = accuracy_score(y_test, y_pred_knn)
print("\nK-Nearest Neighbors:")
print("Confusion Matrix:\n", cm_knn)
print("Accuracy:", accuracy_knn)
print(classification_report(y_test, y_pred_knn))
````

Naïve Bayes


````python
# Naïve Bayes
nb_classifier = GaussianNB()
nb_classifier.fit(X_train, y_train)
y_pred_nb = nb_classifier.predict(X_test)

# Evaluation
cm_nb = confusion_matrix(y_test, y_pred_nb)
accuracy_nb = accuracy_score(y_test, y_pred_nb)
print("\nNaïve Bayes:")
print("Confusion Matrix:\n", cm_nb)
print("Accuracy:", accuracy_nb)
print(classification_report(y_test, y_pred_nb))
````

Decision Tree


````python
# Decision Tree
dt_classifier = DecisionTreeClassifier(criterion='entropy', random_state=0)
dt_classifier.fit(X_train, y_train)
y_pred_dt = dt_classifier.predict(X_test)

# Evaluation
cm_dt = confusion_matrix(y_test, y_pred_dt)
accuracy_dt = accuracy_score(y_test, y_pred_dt)
print("\nDecision Tree:")
print("Confusion Matrix:\n", cm_dt)
print("Accuracy:", accuracy_dt)
print(classification_report(y_test, y_pred_dt))
````

Model Comparison and Selection

````python
# Compare model performances
model_comparison = pd.DataFrame({
    'Model': ['Logistic Regression', 'K-Nearest Neighbors', 'Naïve Bayes', 'Decision Tree'],
    'Accuracy': [accuracy_logistic, accuracy_knn, accuracy_nb, accuracy_dt],
    'Errors': [
        cm_logistic[0,1] + cm_logistic[1,0],
        cm_knn[0,1] + cm_knn[1,0],
        cm_nb[0,1] + cm_nb[1,0],
        cm_dt[0,1] + cm_dt[1,0]
    ]
})

print("\nModel Comparison:")
print(model_comparison)

# Identify the best model
best_model = model_comparison.loc[model_comparison['Accuracy'].idxmax()]
print("\nBest Model:")
print(best_model)
````

Prediction with Best Model


````python
# Using the best model to make predictions (example with new data)
# Let's assume we want to predict for a student with reading score=70 and writing score=80
sample_data = sc.transform([[70, 80]])

# Predict with all models for comparison
print("\nPrediction for student with reading=70, writing=80:")
print("Logistic Regression:", "Pass" if logistic_classifier.predict(sample_data)[0] == 1 else "Fail")
print("KNN:", "Pass" if knn_classifier.predict(sample_data)[0] == 1 else "Fail")
print("Naïve Bayes:", "Pass" if nb_classifier.predict(sample_data)[0] == 1 else "Fail")
print("Decision Tree:", "Pass" if dt_classifier.predict(sample_data)[0] == 1 else "Fail")
````

 **18- Perform the cluster analysis such as K-means and Horizontal for any field from the chosen dataset.**

Part1 : k means

````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
````

````python
dataset = pd.read_csv('StudentsPerformance.csv')
X = dataset.iloc[:, [5, 6, 7]].values  # math score, reading score, writing score
print(X)
````

````python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=1)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
````

````python
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=1)
y_kmeans = kmeans.fit_predict(X)
print(y_kmeans)
````

````python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.scatter(X_pca[y_kmeans == 0, 0], X_pca[y_kmeans == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(X_pca[y_kmeans == 1, 0], X_pca[y_kmeans == 1, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(X_pca[y_kmeans == 2, 0], X_pca[y_kmeans == 2, 1], s=100, c='green', label='Cluster 3')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='yellow', label='Centroids (raw)')

plt.title('Clusters of Students (PCA Projection)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
````
